<h2>
    <b>
        5. Limites de funções
    </b>
</h2>

O SymPy oferece a função <b>limit</b> para lidar com limites de funções simbólicas, então dado o limite:

$$
\lim_{x \to a} f(x) = L
$$

a estrutura desse comando é:

<hr>
 <center><b>L = <b> limit(</b> f(x)<b>,</b> x<b>,</b> a<b>,</b> lado)</b></center>
<hr>

<font color='#008000'><b>Exemplo 9: </b></font> Determine as assíntotas verticais e horizontais, se houver, da função $h(x)$ definida por:

$$ h(x) = \frac{3x}{\sqrt{2x^2-4x}} $$

&nbsp; <b>Passo 1</b>: Declarar a função $h(x)$.

````python
h = 3*x/sqrt(2*x**2-4*x)
````

&nbsp; <b>Passo 2</b>: Utilizar comando <b>solve</b> para resolver $ 2x^2-4x = 0$ e determinar os valores de $x$ para os quais a função $h$ não está definida.

````python
pontos = solve(2*x**2-4*x,x)
pontos
````

````python
p1 = pontos[0]
p1
````

````python
p2 = pontos[1]
p2
````

&nbsp; <b>Passo 3</b>: Utilizar comando <b>limit</b> para determinar a assíntota vertical em <b>p1</b>.

````python
limit(h,x,p1,'-')
````

&nbsp; <b>Passo 4</b>: Utilizar comando <b>limit</b> para determinar a assíntota vertical em <b>p2</b>.

````python
limit(h,x,p2,'+')
````

&nbsp; <b>Passo 5</b>: Utilizar comando <b>limit</b> para determinar as assíntotas horizontais .


````python
h1 = limit(h,x,oo)
h1
````

````python
h2 = limit(h,x,-oo)
h2
````

&nbsp; <b>Passo 6</b>: Utilizar comando <b>plot</b> para exibir $h(x)$ .

````python
plot(h,h1,h2,legend=True,ylim=[-3,10])
````

<hr>
<font color="red"><b>ATENÇÃO:</b></font>

Note que SymPy calcula erroneamente o limite:

````python
limit(h,x,p1,"+")
````

Não faz sentido estudarmos este limite, pois os valores muito pequenos à direita de $0$ não pertencem ao domínio de $h$. Portanto, ao utilizar o SymPy para estudar funções com pontos de descontinuidade, é necessário verificar se o resultado faz sentido matematicamente.

<h2><b>
6. Derivadas de funções
</b></h2>

O comando <b>diff</b> é uma funcionalidade do SymPy capaz de calcular derivadas de funções simbólicas. A derivada

$$  \frac{d^n }{d x^n}\left [ f(x) \right ] $$

é escrita de forma equivalente em Python:

<hr>
 <center><b>diff( <b>f(x)<b> ,</b> x<b> ,</b> n<b> )</b></center>
<hr>

Se $n=1$ então atribuir a ordem torna-se opcional:
<hr>
 <center><b>diff(<b> f(x)<b> ,</b> x<b> )</b></center>
<hr>

<font color='#008000'><b>Exemplo 10: </b></font> Calcule $f'(x)$, se $f(x)=x^2$.

````python
f = x**2
````

&nbsp; <b>Passo 2</b>: Utilizar comando <b>diff</b> para calcular a derivada.

````python
diff(f,x)
````

<h2><b>
6.2 Derivação implícita
</b></h2>

Utiliza-se a função <b>idiff</b> (abreviação para <i>implicit differentiation</i>) para derivar implicitamente uma equação $F(x,y)=0$. Em Python, a derivada $\frac{dy}{dx}$ de $F(x,y)=0$ é escrita de forma equivalente:



<hr>
 <center><b>idiff(</b> &lt;expressão ( F(x,y) )&gt;<b>,</b>&lt;variável ( y )&gt;<b>,</b>&lt;variável ( x )&gt;<b>)</b></center>
<hr>

<b>Lê-se:</b> Obter a derivada $\frac{dy}{dx}$ de $F(x,y)$.

&nbsp; <b>Passo 1</b>: Declarar variáveis.

````python
x, y = symbols('x y')
````

&nbsp; <b>Passo 2</b>: Declarar $F(x,y)$.

````python
F = x**2+y**2-25
F
````

&nbsp; <b>Passo 3</b>: Utilizar a função <b>idiff</b> para calcular a derivada implícita.

````python
dydx = idiff(F,y,x)
dydx
````
